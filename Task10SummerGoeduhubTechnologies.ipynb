***One-Hot Encoding***<br>
For categorical variables where no such ordinal relationship exists, the integer encoding is not enough.<br>

In fact, using this encoding and allowing the model to assume a natural ordering between categories may result in poor performance or unexpected results (predictions halfway between categories).<br>

In this case, a one-hot encoding can be applied to the integer representation. This is where the integer encoded variable is removed and a new binary variable is added for each unique integer value.<br>

In the color variable example, there are 3 categories and therefore 3 binary variables are needed. A 1 value is placed in the binary variable for the color and 0 values for the other colors.<br>

***Dummy Variable Trap in Regression Models***

Using categorical data in Multiple Regression Models is a powerful method to include non-numeric data types into a regression model. Categorical data refers to data values which represent categories - data values with a fixed and unordered number of values, for instance gender (male/female) or season (summer/winder/spring/fall). In a regression model, these values can be represented by dummy variables - variables containing values such as 1 or 0 representing the presence or absence of the categorical value.

***Salary Dataset of 52 professors having categorical columns. Apply dummy variables concept and one-hot-encoding on categorical columns.***

***This is a very small data set consisting of salary data for 52 professors at a small college, categorized by gender, professor rank, highest degree, and years of service paired with salary.*** 

***Dataset Link***<br>- ***https://data.princeton.edu/wws509/datasets/salary.dat***

```python
#import the required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_table("https://data.princeton.edu/wws509/datasets/salary.dat", delim_whitespace = True)
df.head()
```

```python
df['sx'].unique()
```

```python
# Turn my column into a dummy value
# By dropping the first column we did not lose any information, right?
dummy1 = pd.get_dummies(df['sx'], drop_first=True)
# Take a look
dummy1.head()
```

```python
# Now let's cancatenate everything...
df = pd.concat([df, dummy1], axis=1).drop('sx', axis=1)
df.head()
```

```python
# Rearranging column order (optional)
cols = df.columns.tolist()
# Indexing & Slicing Techniques
cols = cols[-1:] + cols[:-1]
cols
```

```python
df = df[cols]
df.head()
```

```python
# Now let's deal with rank (rk):
df['rk'].unique()
```

```python
# Turn my column into a dummy value
dummy2 = pd.get_dummies(df['rk'])
# Take a look
dummy2.head()
```

```python
df = pd.concat([df, dummy2], axis=1).drop('rk', axis=1)
df.head()
```

```python
# Now the last one (dg = Highest degree, coded 1 if doctorate, 0 if masters):
df['dg'].unique()
```

```python
# Turn my column into a dummy value
dummy3 = pd.get_dummies(df['dg'])
# Take a look
dummy3.head()
```

```python
# let's simplify it once more
dummy3 = dummy3.drop('masters', axis=1)
dummy3.head()
```

```python
# Concatenating now...And finally let's go to learn machine learning ... 
df = pd.concat([df, dummy3], axis=1).drop('dg', axis=1)
df.head()
```
